```python
import pandas as pd
import glob
path = r'/Users/balance/Desktop/ARK Investment' # use your path
all_files = glob.glob(path + "/*.csv")
li = []
for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)
frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
# change column name for easier to process data

# frame.weight(%).describe()

frame1=frame.rename(columns={'weight(%)':'weight_percentage'})
```

```python
frame1.weight_percentage.describe()
```

```python
### 4.	Categorized the stocks into 3 different level base on weight percentage. (‘Buy’, ‘Consider’, ‘Not_Consider’)

bin_edges = [1,3,5,12]
bin_names = ['Not_Consider', 'Consider', 'Buy']

frame1['Holding']=pd.cut(frame1['weight_percentage'],bin_edges,labels=bin_names)
frame1
```

```python
Buy=frame1.query('Holding=="Buy"')
```
